## Mixture Gaussian Apllication ... 

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter

# the random data
x1 = np.random.randn(1000)
y1 = np.random.randn(1000)

x2 = np.random.randn(1000)
y2 = np.random.randn(1000) + 5

x = np.r_[x1,x2]
y = np.r_[y1,y2]

nullfmt   = NullFormatter()         # no labels

# definitions for the axes
left, width = 0.1, 0.65
bottom, height = 0.1, 0.65
bottom_h = left_h = left+width+0.02

rect_scatter = [left, bottom, width, height]
rect_histx = [left, bottom_h, width, 0.2]
rect_histy = [left_h, bottom, 0.2, height]

# start with a rectangular Figure
plt.figure(1, figsize=(8,8))

axScatter = plt.axes(rect_scatter)
axHistx = plt.axes(rect_histx)
axHisty = plt.axes(rect_histy)

# no labels
axHistx.xaxis.set_major_formatter(nullfmt)
axHisty.yaxis.set_major_formatter(nullfmt)

# the scatter plot:
axScatter.scatter(x, y)

# now determine nice limits by hand:
binwidth = 0.25
xymax = np.max( [np.max(np.fabs(x)), np.max(np.fabs(y))] )
lim = ( int(xymax/binwidth) + 1) * binwidth

axScatter.set_xlim( (-lim, lim) )
axScatter.set_ylim( (-lim, lim) )

bins = np.arange(-lim, lim + binwidth, binwidth)
axHistx.hist(x, bins=bins)
axHisty.hist(y, bins=bins, orientation='horizontal')

axHistx.set_xlim( axScatter.get_xlim() )
axHisty.set_ylim( axScatter.get_ylim() )

plt.show()
```

# 3D plot ...

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p)*np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)
```

## Preparing Data

```python
phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y).T
```

# pcolor ...

## Denoise with FFT ... 

```python
im_array_t = np.fft.fftshift(np.fft.fft2(im_array))
im_array_n_t = np.fft.fftshift(np.fft.fft2(im_array_n))
kx = np.arange(1,im_array_n_t.shape[1]+1,1)
ky = np.arange(1,im_array_n_t.shape[0]+1,1)
KX,KY = np.meshgrid(kx,ky)
GaussFilter = np.exp(-0.0001*(KX - im_array_n_t.shape[1] / 2)**2 - 0.0001*(KY - im_array_n_t.shape[0] / 2)**2)

im_array_n_t_F = np.multiply(im_array_n_t,GaussFilter)
im_array_n_t_F_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_n_t_F)).real
im_array_t_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_t)).real
im_array_n_t_ifft = np.fft.ifft2(np.fft.ifftshift(im_array_n_t)).real
```

# Min-Max Problem:

## $$ \begin{eqnarray} \min_{x_1,x_2} \max( x_1 & , & x_2 ) \\ 2x_1 & + & x_2 & \geq 1 \\ x_1 & + & 3x_2 & \geq 1 \\ x_1 &&& \geq 0 \\ x_2 &&& \geq 0 \\ \end{eqnarray}$$

```python
xx1 = np.arange(0, 1, 0.001)
xx2 = np.arange(0, 1, 0.001)
XX1, XX2 = np.meshgrid(xx1, xx2)

obj_fn =  np.maximum(XX1, XX2)
c1_fn = 2 * XX1 +  XX2
c2_fn = XX1 + 3 * XX2

c_region = obj_fn.copy()

c_region[(c1_fn > 1) & (c2_fn > 1)] = 200
c_region[(c1_fn < 1) | (c2_fn < 1)] = 255


fig = pylab.figure()

c2_fn_plt = pylab.imshow(c_region,origin=[0,0],cmap=cm.Greys_r)
pylab.xticks(np.linspace(0,len(xx1),11),np.linspace(0,1,11))
pylab.yticks(np.linspace(0,len(xx2),11),np.linspace(0,1,11))


obj_fn_values = np.arange(0,0.5,0.03)
obj_fn_plt = pylab.contour(obj_fn,obj_fn_values,colors='g')
pylab.clabel(obj_fn_plt, inline=1, fontsize=10)

# ax, = fig.get_axes()
# ax.text(0.8*len(xx1), 0.05*len(xx1), r"$x_1 + 3x_2 = 1$", fontsize=20, color="b")
# ax.text(0.05*len(xx1), 0.9*len(xx1), r"$2x_1 + x_2 = 1$", fontsize=20, color="b")
# ax.text(0.8*len(xx1), 0.9*len(xx1), r"$2x_1 + x_2 \geq 1$", fontsize=20)
# ax.text(0.8*len(xx1), 0.8*len(xx1), r"$x_1 + 3x_2 \geq 1$", fontsize=20)

fig.savefig("MinMax_Demo.png",figsize=(20,15),dpi=200)
```
